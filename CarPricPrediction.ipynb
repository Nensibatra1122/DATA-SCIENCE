Checking if the year column set has numeric or string datatype values

Changing the year string datatype value into integer

```python
import pandas as pd
import numpy as np

# Assuming 'year' column may contain non-string values
# Convert 'year' to string type
car['year'] = car['year'].astype(str)

# Replace empty strings ('') with NaN
car['year'] = car['year'].replace('', np.nan)

# Remove non-numeric characters like 'k' from the 'year' column
car['year'] = car['year'].str.replace(r'\D', '', regex=True)

# Convert the 'year' column to integers and replace NaN and empty strings with 0
car['year'] = car['year'].fillna(0).astype(int)

```

Finding the model with a random state of TrainTestSplit where the model was found to give almost 0.92 as r2_score

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
pipe.predict(pd.DataFrame(columns=X_test.columns,data=np.array(['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']).reshape(1,5)))
```

The best model is found at a certain random state

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
